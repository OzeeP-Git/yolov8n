```python
yaml_content = """
path: /content/VehiclesDetectionDataset

train: images/train
val: images/valid
test: images/test

names: ['Ambulance', 'Bus', 'Car', 'Motorcycle', 'Truck']
"""

with open('/content/VehiclesDetectionDataset/dataset.yaml', 'w') as f:
    f.write(yaml_content.strip())

print("✅ dataset.yaml updated successfully.")
```

```python
import time

start_time = time.time()

!yolo task=detect \
    mode=train \
    model=yolov8n.pt \
    data=/content/VehiclesDetectionDataset/dataset.yaml \
    epochs=50 \
    imgsz=640 \
    batch=16 \
    name=YOLOv8-vehicles

end_time = time.time()
print(f"\n✅ Training completed in {((end_time - start_time)/60):.2f} minutes.")
```
